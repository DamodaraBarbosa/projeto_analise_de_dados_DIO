```python
import pandas as pd
```

```python
# pode-se selecionar determinadas linhas do dataframe fazendo uma espécie de fatiamento

df[0:10]
```

```python
# caso queira selecionar uma linha em específico é possível utilizar o método .iloc[]
# o que retorna uma Series

df.iloc[32]
```

```python
# caso queira retornar um data frame deve-se colocar o index em colchetes duplos

df.iloc[[32]]
```

```python
type(df.iloc[[32]])
```

```python
# para obter a quantidade de linhas e colunas

df.shape
```

```python
# para obter apenas os nomes das colunas

df.columns
```

```python
# uma outra maneira é o uso do método info() que além de mostrar o tipo de dado em cada coluna também mostra a quantidade
# de linhas e colunas, assim como o nome das colunas, e os valores não nulos em cada coluna

df.info()
```

```python
# uma forma inteligente de mostrar as últimas linhas do dataframe é utilizar o método tail()

df.tail()
```

```python
# o método describe() retorna estatísticas descritivas num formato resumido, porém bastante completo

df.describe()
```
