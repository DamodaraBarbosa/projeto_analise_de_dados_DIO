```python
# como localizar um determinado continente:

# 1¬∫ - como mostrado na aula:

oceania = df.loc[df['Continente'] == 'Oceania']
df.head()

```

```python
# 2¬∫ - outra maneira, que eu utilizo mais

oceania = df[df['Continente'] == 'Oceania']
oceania.head()
```

```python
# 3¬∫ - m√©todo que eu mais utilizo e que acho mais elegante, que utilizando o m√©todo query()

oceania = df.query('Continente == "Oceania"')
oceania.head()
```

```python
oceania['Continente'].unique()
```

```python
# caso queira fazer agrupamentos:

grupos_continentes = df.groupby(by= 'Continente')
```

```python
# para definir a quantidade de pa√≠ses por continente:
# usando m√©todo nunique()

grupos_continentes[['Pa√≠s']].nunique()
```

```python
# pode-se gerar um dataframe com a m√©dia da expectativa de vida em cada ano

df.groupby(by= 'Ano')[['Expectativa de vida']].mean().sample(10)
```

```python
# a m√©dia da expectativa de vida por pa√≠s:

expectativa_por_pais = df.groupby(by= 'Pa√≠s')[['Expectativa de vida']].mean()
expectativa_por_pais.head()
```

```python
# caso queira procurar o Brasil

expectativa_por_pais.query('Pa√≠s == "Brazil"')
```
