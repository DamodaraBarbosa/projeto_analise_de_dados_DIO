```python
# para identificar os valores na coluna 'Continente':

df['Continente'].unique()
```

```python
# como localizar um determinado continente:

# 1º - como mostrado na aula:

oceania = df.loc[df['Continente'] == 'Oceania']
df.head()

```

```python
# 2º - outra maneira, que eu utilizo mais

oceania = df[df['Continente'] == 'Oceania']
oceania.head()
```

```python
# 3º - método que eu mais utilizo e que acho mais elegante, que utilizando o método query()

oceania = df.query('Continente == "Oceania"')
oceania.head()
```

```python
oceania['Continente'].unique()
```

```python
# caso queira fazer agrupamentos:

grupos_continentes = df.groupby(by= 'Continente')
```

```python
# para definir a quantidade de países por continente:
# usando método nunique()

grupos_continentes[['País']].nunique()
```

```python
# pode-se gerar um dataframe com a média da expectativa de vida em cada ano

df.groupby(by= 'Ano')[['Expectativa de vida']].mean().sample(10)
```

```python
# a média da expectativa de vida por país:

expectativa_por_pais = df.groupby(by= 'País')[['Expectativa de vida']].mean()
expectativa_por_pais.head()
```

```python
# caso queira procurar o Brasil

expectativa_por_pais.query('País == "Brazil"')
```
