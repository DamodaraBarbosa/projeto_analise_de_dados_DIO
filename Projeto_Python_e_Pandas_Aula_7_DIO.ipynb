Importando as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Gerando o dataframe

```python
df = pd.read_excel('/content/AdventureWorks.xlsx')
```

```python
df.head()
```

```python
# dimensões do dataframe:

df.shape
```

```python
df.info()
```

```python
# valor da receita total:

df['Valor Venda'].sum()
```

```python
# custo total:

df['Custo'] = df['Custo Unitário'] * df['Quantidade']
```

```python
df.head()
```

```python
# custo total:
df['Custo'].sum()
```

```python
# agora para saber o lucro:

df['Lucro'] = df['Valor Venda'] - df['Custo']
df.head()
```

```python
# lucro total:

round(df['Lucro'].sum(), 2)
```

```python
# tempo para envio do produto:

df['Envio'] = df['Data Envio'] - df['Data Venda']
df.head()
```

```python
df['Envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
df.head()
```

```python
df['Envio'].dtypes
```

```python
# tempo de envio médio por marca:

round(df.groupby('Marca')['Envio'].mean(), 2)
```

```python
# valores nulos:

df.isna().sum()
```

Lucro por ano e por marca

```python
# adicionando coluna com ano da venda:

df['Ano'] = df['Data Venda'].dt.year
df.head(1)
```

```python
lucro_ano = pd.DataFrame(df.groupby(['Ano', 'Marca'])['Lucro'].sum())
lucro_ano
```

```python
# resetando o index:
lucro_ano.reset_index()
```

```python
# quantidade vendida de cada produto:

qtde_produtos_vendidos = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending= False)
qtde_produtos_vendidos
```

```python
sns.set()
plt.barh(qtde_produtos_vendidos.keys(), qtde_produtos_vendidos.values)
plt.ylabel('Produto')
plt.xlabel('Vendas')
plt.show()
```

```python
# lucro por ano:

lucro_por_ano = df.groupby(df['Data Venda'].dt.year)['Lucro'].sum()
lucro_por_ano.plot.bar()
plt.title('Ano x Venda')
plt.xlabel('Ano')
plt.ylabel('Venda')
plt.show()
```

```python
# selecionando apenas as vendas de 2009:

venda_2009 = df.query('Ano == 2009')
venda_2009.head()
```

```python
# gráfico Lucro x Mês:

mes_lucro_2009 = venda_2009.groupby(venda_2009['Data Venda'].dt.month)['Lucro'].sum()
mes_lucro_2009
```

```python
# plotando em um gráfico:

plt.plot(mes_lucro_2009.keys(), mes_lucro_2009.values)
plt.title('Lucro x Mês')
plt.xlabel('Mês')
plt.ylabel('Lucro')
plt.show()
```

```python
# lucro no ano 2009 por marca:

marca_lucro_2009 = venda_2009.groupby('Marca')['Lucro'].sum()
marca_lucro_2009
```

```python
plt.bar(marca_lucro_2009.keys(), marca_lucro_2009.values)
plt.show()
```

```python
# lucro por classe no ano 2009:

lucro_classe_2009 = venda_2009.groupby('Classe')['Lucro'].sum()
lucro_classe_2009
```

```python
plt.bar(lucro_classe_2009.keys(), lucro_classe_2009.values)
plt.show()
```
