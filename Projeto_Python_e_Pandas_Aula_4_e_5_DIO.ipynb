```python
import pandas as pd
```

Importando os arquivos

```python
aracaju = pd.read_excel('/content/drive/MyDrive/BOOTCAMP DIO/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Aracaju.xlsx')
fortaleza = pd.read_excel('/content/drive/MyDrive/BOOTCAMP DIO/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Fortaleza.xlsx')
salvador = pd.read_excel('/content/drive/MyDrive/BOOTCAMP DIO/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Salvador.xlsx')
recife = pd.read_excel('/content/drive/MyDrive/BOOTCAMP DIO/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Recife.xlsx')
natal = pd.read_excel('/content/drive/MyDrive/BOOTCAMP DIO/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Natal.xlsx')
```

```python
# concatenando os dataframes

df = pd.concat([aracaju, fortaleza, salvador, recife, natal])
```

```python
# o tipo de dado de cada coluna 

df.dtypes
```

```python
# alterando o tipo de dado de cada coluna

df['LojaID'] = df['LojaID'].astype('object')
```

```python
df.dtypes
```

```python
# verificando valores faltantes

df.isnull().sum()
```

```python
# para preencher dados nulos, quando estes existirem:
# preenchendo com a média:

df['Vendas'].fillna(df['Vendas'].mean(), inplace= True)
```

```python
# preenchendo com zero:

df['Vendas'].fillna(0, inplace= True)
```

```python
# apagando os dados nulos:

df.dropna(inplace= True)
```

```python
# apagando dados nulos de uma determinada coluna:

df.dropna(subset= ['Vendas'], inplace= True)
```

```python
# só apaga se houver dados faltantes em todas as colunas:

df.dropna(how= 'all', inplace= True)
```

Criando novas colunas

```python
df['Receita'] = df['Vendas'] * df['Qtde']
```

```python
df.head()
```

```python
# retornando maior receita:

df['Receita'].max()
```

```python
# ou 

max(df['Receita'])
```

```python
# retornando menor receita:

df['Receita'].min()
```

```python
# ou 

min(df['Receita'])
```

```python
# definindo a coluna receita/vendas:

df['Receita/vendas'] = df['Receita']/df['Vendas']
```

```python
df.head()
```

```python
# uso do método nlargest()

df.nlargest(3, 'Receita')
```

```python
# uso do método nsmalleste()

df.nsmallest(3, 'Receita')
```

```python
# agrupando dados por cidade:

cidades_group = df.groupby(by= 'Cidade')['Receita'].sum()
cidades_group
```

## Trabalhando com datas

```python
# transformando os dados da coluna 'Data' para int:

df['Data'] = df['Data'].astype('int')
```

```python
df.head()
```

```python
# passando a coluna 'Data' novamente para o tipo datetime

df['Data'] = pd.to_datetime(df['Data'])
```

```python
df.head()
```

```python
# agrupamento de receita por ano:

df.groupby(df['Data'].dt.year)['Receita'].sum()
```

```python
# criar coluna ano:

df['Ano_venda'] = df['Data'].dt.year
df.sample(5)
```

```python
# extrair mês e dia da venda

df['Mês_venda'] = df['Data'].dt.month
df.sample(5)
```

```python
df['Dia_venda'] = df['Data'].dt.day
```

```python
df.sample(5)
```

```python
# retornando data mínima:

dt_min = df['Data'].min()
```

```python
# retornando data máxima:

dt_max = df['Data'].max()
```

```python
# calculando a diferença entre as datas:

dt_max - dt_min
```

```python
# criando coluna para trimestre:

df['Trimestre'] = df['Data'].dt.quarter
```
